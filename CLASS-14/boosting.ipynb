# XGBoost and CatBoost (Class-14, Module-1)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv('/content/drive/MyDrive/Ensemble Models/Sleep_health_dataset.csv')
data.head()
```

```python
data = data.drop('Person ID', axis=1)
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data = data.drop('Blood Pressure', axis=1)
```

```python
to_combine = ['Sleep Apnea', 'Insomnia']
data['Sleep Disorder'] = data['Sleep Disorder'].replace(to_combine, 'Sleep Disorder')
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df = data.copy()
```

```python
enc = LabelEncoder()
```

```python
df.Gender = enc.fit_transform(df.Gender)
df.Occupation = enc.fit_transform(df.Occupation)
df['BMI Category'] = enc.fit_transform(df['BMI Category'])
df['Sleep Disorder'] = enc.fit_transform(df['Sleep Disorder'])
```

```python
df
```

```python
df.columns
```

```python
features = ['Gender', 'Age', 'Occupation', 'Sleep Duration', 'Quality of Sleep',
       'Physical Activity Level', 'Stress Level', 'BMI Category', 'Heart Rate',
       'Daily Steps']

x = df[features]
y = df['Sleep Disorder']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=10)
```

# XGBoost

```python
from xgboost import XGBClassifier
xgb = XGBClassifier(random_state=10, max_depth=12,n_estimators=300)
```

```python
xgb.fit(x_train,y_train)
xgb_pred = xgb.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
```

```python
confusion_matrix(y_test, xgb_pred)
```

```python
print('accuracy:',accuracy_score(y_test, xgb_pred))
print('precision:',precision_score(y_test,xgb_pred))
print('recall:',recall_score(y_test, xgb_pred))
print('F1-score:', f1_score(y_test, xgb_pred))
```

```python
X = data[['Gender', 'Age', 'Occupation', 'Sleep Duration', 'Quality of Sleep',
       'Physical Activity Level', 'Stress Level', 'BMI Category', 'Heart Rate',
       'Daily Steps']]
Y = data['Sleep Disorder']
```

```python
from catboost import CatBoostClassifier
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=10)
```

```python
params = {'learning_rate': 0.1, 'depth': 6, 'iterations': 100}

# iterations is the max number of trees that can be built and learning rate is to adjust the gradient step.
```

```python
cat = CatBoostClassifier(**params)
```

```python
cat_var = np.where(X_train.dtypes != np.float)[0]
```

```python
cat.fit(X_train, Y_train, cat_features=cat_var, plot=True)
```

```python
cat_pred = cat.predict(X_test)
```

```python
confusion_matrix(Y_test, cat_pred)
```

```python
accuracy_score(Y_test, cat_pred)
```
