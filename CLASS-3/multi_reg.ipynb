```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
sns.set()
import statsmodels.api as sm 
```

# 2 independent variables

```python
data = pd.read_csv('TV_Radio.csv')
data.head()
```

```python
data.shape
```

```python
data.describe()
```

```python
y = data['Sales']
x1 = data[['TV', 'Radio']]
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
```

```python
results.summary()
```

```python
df = pd.read_csv('advertising.csv')
df.head()
```

```python
y = df['Sales']
x1 = df[['TV', 'Radio', 'Newspaper']]
```

```python
x = sm.add_constant(x1)

new_res = sm.OLS(y,x).fit()
```
