# Market Segmentation

```python
import pandas as pd
import numpy as np
import seaborn as sns
sns.set()
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('market_segmentation.csv')
data.head()
```

```python
data.shape
```

```python
plt.scatter(data['Satisfaction'],data['Loyalty'])
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
plt.show()
```

```python
x = data.copy()
```

```python
kmeans = KMeans(2)
kmeans.fit(x)
```

```python
clusters = x.copy()
clusters['cluster_pred']=kmeans.fit_predict(x)
```

```python
plt.scatter(clusters['Satisfaction'],clusters['Loyalty'],c=clusters['cluster_pred'],cmap='rainbow')
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
plt.show()
```

```python
#standardizing

from sklearn import preprocessing
x_scaled = preprocessing.scale(x)
x_scaled
```

```python
wcss = []
for i in range (1,10):
    kmeans = KMeans(i)
    kmeans.fit(x_scaled)
    wcss.append(kmeans.inertia_)
    
wcss
```

```python
plt.plot(range(1,10),wcss)
plt.xlabel('no. of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
kmeans_new = KMeans(4)
kmeans_new.fit(x_scaled)
clusters_new = x.copy()
clusters_new['cluster_pred'] = kmeans_new.fit_predict(x_scaled)
```

```python
clusters_new.head()
```

```python
plt.scatter(clusters_new['Satisfaction'],clusters_new['Loyalty'],c=clusters_new['cluster_pred'],cmap='rainbow')
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
plt.show()
```

* Group - 1: extremely low loyalty. the low satisfaction might explain this.
* Group-2: medium to highly satisfied. and hence are moderately loyal.
* Group-3: low to moderately satisfied, however they are loyal. These customers are a special case and are low in number.
* Group-4: **target audiance** they are highly satisfied and hence highly loyal.
