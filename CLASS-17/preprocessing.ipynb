```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv('Sleep_health_dataset.csv')
data.head()
```

```python
data = data.drop('Person ID', axis=1)
data.head()
```

```python
data['Sleep Disorder'] = data['Sleep Disorder'].fillna('None')
data
```

```python
data.info()
```

```python
to_combine = ['Normal', 'Normal Weight']
data['BMI Category'] = data['BMI Category'].replace(to_combine, 'Normal')
```

```python
data = pd.concat([data, data['Blood Pressure'].str.split('/', expand=True)], axis=1).drop('Blood Pressure', axis=1)
data = data.rename(columns={0: 'bp_upper', 1: 'bp_lower'})
data['bp_upper'] = data['bp_upper'].astype('int')
data['bp_lower'] = data['bp_lower'].astype('int')
```

```python
data.info()
```

```python
from sklearn.preprocessing import LabelEncoder

enc = LabelEncoder()
```

```python
df_transformed = data.copy()
df_transformed.head()
```

```python
df_transformed['Occupation'] = enc.fit_transform(data['Occupation'])
df_transformed['BMI Category'] = enc.fit_transform(data['BMI Category'])
df_transformed['Sleep Disorder'] = enc.fit_transform(data['Sleep Disorder'])
df_transformed['Gender'] = enc.fit_transform(data['Gender'])
df_transformed.head()
```

```python
df_transformed.columns
```

```python
columns_reordered = ['Gender', 'Age', 'Occupation', 'Sleep Duration', 'Quality of Sleep',
       'Physical Activity Level', 'Stress Level', 'BMI Category', 'Heart Rate',
       'Daily Steps', 'bp_upper', 'bp_lower', 'Sleep Disorder']

df_final = df_transformed[columns_reordered]
df_final
```

```python
df_final.to_csv('data_preprocessed.csv', header=False, index=False)
```
